Now, we will find out the total funded amount based on the purpose of the loan

```python
#  Segmented analysis on the purpose of the loan based on total funded amount
total_funded_amt_basedon_purpose = loan.groupby('purpose').funded_amnt.sum().sort_values(ascending=False)
total_funded_amt_basedon_purpose
```

Now, we will find out the total frequency of loans based on the purpose of the loan

```python
#  Segmented analysis on the purpose of the loan based on total No of applications
count_basedon_purpose = loan.groupby('purpose').funded_amnt.count().sort_values(ascending=False)
count_basedon_purpose
```

```python
#Initail Bivarate analysis
# Compute the correlation matrix
corr = loan.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(12, 10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5,annot=True, cbar_kws={"shrink": .5})
print("the most strongest relation are ")
print("1.Loan amount and funded amount")
print("2.Funded amount and investment")
print("3.Funded amount and Funded amount investemnt")
```

```python
#Initial bivariate analysis  
print("Visualizing Multidimensional Relationships")
plt.figure(figsize=(45,10))
g = sns.PairGrid(loan, vars=['loan_amnt', 'annual_inc', 'dti'],
                 hue='purpose', height=4.5, aspect=1)
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

```python
#  Segmented analysis on the purpose of the loan based on total funded amount
funded_amt_purpose = pd.concat([total_funded_amt_basedon_purpose, count_basedon_purpose], axis=1,sort=False)
funded_amt_purpose.columns = ['total_funded_amt_basedon_purpose', 'count_basedon_purpose']
total = loan['funded_amnt'].sum()
funded_amt_purpose['Percentage'] = round((funded_amt_purpose['total_funded_amt_basedon_purpose']*100/total),2)
funded_amt_purpose
```

```python
#percentage of Defaulter segmented on employment lenght
Loan_emp_yers=loan[loan['loan_status']=='Charged Off']
Loan_emp_yers_total=loan.groupby('emp_length')['loan_status'].count().sort_values(ascending=False).reset_index()
#print(Loan_emp_yers_total)
Loan_emp_yers=Loan_emp_yers.groupby('emp_length')['loan_status'].count().sort_values(ascending=False).reset_index()
Loan_emp_yers['loan_status']=Loan_emp_yers['loan_status'].astype(int)

Loan_emp_yers['Defaluter per']=Loan_emp_yers['loan_status']*100/Loan_emp_yers_total['loan_status']
Loan_emp_yers['Defaluter per']=Loan_emp_yers['Defaluter per'].astype(int)
Loan_emp_yers['emp_length']=Loan_emp_yers['emp_length'].astype(int)
Loan_emp_yers=Loan_emp_yers.sort_values(by='Defaluter per',ascending=False).reset_index(drop=True)

plt.figure(figsize=(15,6))
fig=sns.barplot( x="emp_length",y="Defaluter per", data=Loan_emp_yers,order=Loan_emp_yers['emp_length'],palette='Set1')
plt.xlabel("Years Of Employment")
plt.ylabel("% Of Defaulters")
plt.show(fig)
```
